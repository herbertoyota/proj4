```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
site = 'https://data.worldbank.org/country/Japan'
page = requests.get(site)
resposta = page.text
# Inicializa o BeautifulSoup
soup = BeautifulSoup(resposta, 'html.parser')

# Extrai os dados de cada classe
headlines = [item.text.strip() for item in soup.find_all('div', class_='indicator-item__headline-mobile')]
years = [item.text.strip() for item in soup.find_all('div', class_='indicator-item__mobile-above-info-year')]
data_info = [item.text.strip() for item in soup.find_all('div', class_='indicator-item__data-info')]

# Verifica se o número de itens em cada lista é igual
if len(headlines) == len(years) == len(data_info):
    # Cria o DataFrame
    df = pd.DataFrame({
        'Headline': headlines,
        'Year': years,
        'Data Info': data_info
    })
df
```

```python
# List of countries
countries = ['brazil', 'germany', 'japan']  

# List to store the data of the selected countries
datas = []

for country in countries:
    # Create the URL for each country
    site = f'https://data.worldbank.org/country/{country}'
    page = requests.get(site)
    
    if page.status_code == 200:  # Check if the page was loaded successfully
        resposta = page.text
        soup = BeautifulSoup(resposta, 'html.parser')

        # Extract data from specific classes
        headlines = [item.text.strip() for item in soup.find_all('div', class_='indicator-item__headline-mobile')]
        years = [item.text.strip() for item in soup.find_all('div', class_='indicator-item__mobile-above-info-year')]
        data_info = [item.text.strip() for item in soup.find_all('div', class_='indicator-item__data-info')]

        # Check if the number of items is the same in all lists
        if len(headlines) == len(years) == len(data_info):
            # Add data to the set
            for headline, year, info in zip(headlines, years, data_info):
                datas.append({
                    'Country': country.capitalize(),  
                    'Headline': headline,
                    'Year': year,
                    'Data Info': info
                })
# Create a DataFrame with data from all countries
df_countries = pd.DataFrame(datas)
df_countries 
```

```python
# Create a table of specific indicator
gdp_per_capita = df_countries[df_countries['Headline']=='GDP per capita (current US$)']
gdp_per_capita
```
